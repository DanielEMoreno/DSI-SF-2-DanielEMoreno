### 3.2 Clean the columns

```python
iowa_reader = csv.DictReader(open(iowa_file, 'r'), delimiter=',')
header = iowa_reader.next()
data = {k:[] for k in header}

#data_dict = {}
for items in iowa_reader:
    for k in items.items():
        try:
            data[k[0]].append(float(k[1].replace('$', '')))
        except:
            try:
                data[k[0]].append(float(int(k[1])))
            except:
                data[k[0]].append(k[1])

cleaned_iowa_df = pd.DataFrame(data)
```

```python
cleaned_iowa_df['Date'] = pd.to_datetime(cleaned_iowa_df['Date'], format='%m/%d/%Y')

cleaned_iowa_df.info()

cleaned_iowa_df.head()
```

```python
cleaned_iowa_df['Profit'] = ((cleaned_iowa_df['State Bottle Retail'] - cleaned_iowa_df['State Bottle Cost'])
                             *cleaned_iowa_df['Bottles Sold'])
```

### 6.1 Create columns for Q1 2015 sales and Q1 2016 sales.

The owner is not just interested in what locations are currently selling well, but also which areas are showing the most promising sales growth from 2015 to 2016. 

For yearly changes we will focus on the changes from Q1 2015 to Q1 2016 specifically (avoiding effects of season changes).

Note: Q1 is months: January, February, and March.


```python
first_quarter_stores = first_year_stores['Date'] < '2015-03-31'
second_quarter_stores = second_year_stores['Date'] < '2016-03-31'
```

```python
first_quarter_stores_df = pd.DataFrame(first_year_stores[first_quarter_stores])
second_quarter_stores_df = pd.DataFrame(second_year_stores[second_quarter_stores])

q15_sales_by_store = first_quarter_stores_df.groupby(['Store Number'])[['Profit']].apply(np.sum)
q16_sales_by_store = second_quarter_stores_df.groupby(['Store Number'])[['Profit']].apply(np.sum)

cleaned_iowa_df['q15_q16_diff'] = (q16_sales_by_store['Profit'] - q15_sales_by_store['Profit'])
```

```python
cleaned_iowa_df.head()

successful_stores = cleaned_iowa_df[cleaned_iowa_df['q15_q16_diff'] > 0]

successful_stores = successful_stores[successful_stores['County'].map(len) > 1]
#city by county
print len(cleaned_iowa_df)
print len(successful_stores)
```
